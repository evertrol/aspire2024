# Simple demonstration of Aspire package

Set a state first

```python
from aspire import init_circular
state0 = init_circular()
state0
```

Set a duration (1 = one single orbit) and a time step size

```python
duration = 1
tau = 1/12
```

Integrate with Runge Kutta 4

```python
from aspire import integrate_rk4

times, history = integrate_rk4(state0, tau, duration)
times
```

Let's plot this

```python
from aspire.plotting import plot

fig = plot(history)
```

Try for 5 continuous orbits, with different time step sizes.

```python
duration = 5
taus = [1/12, 0.01, 0.001]

fig = None
for tau in taus:
    _, history = integrate_rk4(state0, tau, duration)
    label = rf"$\tau = {tau:6.4f}$"
    if not fig:
        fig = plot(history, label=label)
    else:
        plot(history, ax=fig.gca(), label=label)
fig.gca().legend();
```
